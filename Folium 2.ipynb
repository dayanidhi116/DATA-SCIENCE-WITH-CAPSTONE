```python
!pip install folium
import folium
# Create a map centered on New York City
map_nyc = folium.Map(location=[40.7128, -74.0060], zoom_start=12)
map_nyc
folium.Marker(
    location=[40.7128, -74.0060],
    popup='New York City',
    icon=folium.Icon(color='blue')
).add_to(map_nyc)
map_nyc
folium.CircleMarker(
    location=[40.7128, -74.0060],
    radius=50,
    color='red',
    fill=True,
    fill_color='red',
    fill_opacity=0.6,
    popup='NYC Circle'
).add_to(map_nyc)
map_nyc
```

```python
import pandas as pd
import folium

# Load SpaceX launch CSV data
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
spacex_df = pd.read_csv(url)

# Create a base map centered at average lat/lon
map_spacex = folium.Map(location=[spacex_df['Lat'].mean(), spacex_df['Long'].mean()], zoom_start=5)

# Add markers for each launch site
for idx, row in spacex_df.iterrows():
    folium.Marker(
        location=[row['Lat'], row['Long']],
        popup=f"Launch Site: {row['Launch Site']}",
        icon=folium.Icon(color='red')
    ).add_to(map_spacex)

map_spacex
```

```python
!pip install folium
import folium

# Coordinates for New Delhi, India
india_coords = [28.6139, 77.2090]

# Create a map centered on New Delhi
map_india = folium.Map(location=india_coords, zoom_start=6)

# Add a marker for New Delhi
folium.Marker(
    location=india_coords,
    popup='New Delhi',
    icon=folium.Icon(color='blue')
).add_to(map_india)

# Add a circle marker for New Delhi
folium.CircleMarker(
    location=india_coords,
    radius=50,
    color='red',
    fill=True,
    fill_color='red',
    fill_opacity=0.6,
    popup='New Delhi Circle'
).add_to(map_india)

map_india
```
